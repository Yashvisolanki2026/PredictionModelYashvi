```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#import heart_disease dataset and print first five columns
df = pd.read_csv('heart_disease.csv')
df.head()
```

```python
#shape function to find size
df.shape
```

```python
df.describe()
```

```python
#isnull is used to find no. of null values
print(df.isnull().sum())
```

```python
# Data Cleaning
# Standardize sex values
df['sex'] = df['sex'].str.lower()

# Fill missing values with median
for col in ['trestbps', 'chol', 'thalach']:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
```

```python
x = df.drop('target', axis=1)
y = df['target']
print(x)
```

```python
print(y)
```

```python
# Apply one-hot encoding to categorical features before splitting and scaling
if 'sex' in x.columns:
    x = pd.get_dummies(x, columns=['sex'], drop_first=True)

# Split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Train Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
# Predictions
y_pred = model.predict(x_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", round(accuracy * 100, 2), "%")
```

```python
x_test_pred=model.predict(x_test)
x_test_acc=accuracy_score(x_test_pred,y_test)
print(x_test_acc)
```

```python
 #Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
